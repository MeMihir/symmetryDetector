```python
import cv2
import matplotlib.pyplot as plt
import Symmetry
import numpy as np

%matplotlib inline
```

```python
def display_img(title, img):
    while True:
        cv2.imshow(title, img)
        if cv2.waitKey(5) & 0xFF == 27:
            break
    cv2.destroyAllWindows()
```

```python
sym_detect = Symmetry.SymmetryDetector(threshold=[90,255,0], resolution=180, centroid=(-1,-1), scoreThreshold=0.)
```

```python
img = cv2.imread('./ISIC_0052212.jpg')
display_img('title', img)
plt.imshow(img)
```

```python
output = sym_detect.detectSymmetry(img)
# output
```

```python
plt.figure(figsize=[20,8])
plt.plot(output[:,0], output[:,1])
```
